```python
import pandas as pd
```

```python
df=pd.read_csv("salaries.csv")
df
```

```python
#for drop/remove the columns
input=df.drop("salary_more_then_100k",axis="columns")
input
```

```python
#store salary_more_then_100k columns in target variable
#0 means salary is less than 100k or 1 means salary is more than 100k
target=df["salary_more_then_100k"]
target
```

```python
# loadEncoder
# to convert the label value into numbers
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
# to convert the label value into numbers
input ["company_n"]=le_company.fit_transform(input['company'])
input ["job_n"]=le_job.fit_transform(input['job'])
input ["degree_n"]=le_degree.fit_transform(input['degree'])
input
```

```python
#for remove the columns and store into a new variable
input_n=input.drop(["company","job","degree"],axis="columns")
input_n
```

```python
#load decision tree 
from sklearn import tree
#create model 
model=tree.DecisionTreeClassifier()
```

```python
#train model
#input=input_n or targert vari.=target
model.fit(input_n,target)
```

```python
#for check the model score
model.score(input_n,target)
```

```python
# lets do predicton
#2 for google,1 for computer programer,0 for bachlor acc.to above datasets
#0 means salary is less than 100k or 1 means salary is more than 100k
model.predict([[0,2,1]])
```
